# Generación de nudos aleatorios

Los nudos se generan como:

$$\begin{align*}
    x &= A_{x,1} cos(n_{x,1} t + \phi_{x,1}) + \cdots + A_{x,i} cos(n_{x,i} t + \phi_{x,i}) \\
    y &= A_{y,1} cos(n_{y,1} t + \phi_{y,1}) + \cdots + A_{y,j} cos(n_{y,j} t + \phi_{y,j}) \\
    z &= A_{z,1} cos(n_{z,1} t + \phi_{z,1}) + \cdots + A_{z,k} cos(n_{z,k} t + \phi_{z,k})
\end{align*}$$

Con $t \in S^1$ y $n_x, n_y, n_z \in \mathbb{Z}$. Los parámetros son $i,j,k \in \mathbb{Z}^+$.

```python
from knots_generator import knots
```

Se indican hiperparámetros para generar los nudos:

* points: int, número de puntos que tendrá el nudo.
* noise: float, introducción de ruido blanco en las barras del nudo.
* sc: int, entero más pequeño en la distribución de $n_x, n_y, n_z$.

```python
knot_generator = knots(points=5000, noise=0.001, sc=5)
```

#### Generación de nudos

La función <i>create_random_knot(i,j,k)</i> genera nudos aleatorios (sin barras).

```python
knot_generator.create_random_knot(1,1,2)
print(knot_generator.knot)
knot_generator.plot_knot()
```

#### Generación de nudos con barras

La función <i>create_bars_random_knot(i,j,k)</i> genera nudos aleatorios con barras.

```python
knot_generator.create_bars_random_knot(1,1,2,genus=3)
print(knot_generator.knot)
knot_generator.plot_knot(bars_color='red')
```
